```python
import pandas as pd
```

```python
#importando nosso dataset
data = pd.read_csv('./dataset/GasPricesinBrazil_2004-2019.csv', sep = ';' )
```

```python
data
```

```python
#limpando a coluna unnamed
del data['Unnamed: 0']
```

```python
data.head()
```

```python
#Limpando os dados
data.info()
```

```python
data_pre = data.copy()
```

```python
#Mudando as datas para tipo Datetime
data_pre['DATA INICIAL'] = pd.to_datetime(data_pre['DATA INICIAL'])
data_pre['DATA FINAL'] = pd.to_datetime(data_pre['DATA FINAL'])
```

```python
data_pre.info()
```

```python
#CONVERSAO DE ATIRBUTOS PARA NUMERIC:
for atributo in['MARGEM MÉDIA REVENDA', 'PREÇO MÉDIO DISTRIBUIÇÃO','DESVIO PADRÃO DISTRIBUIÇÃO','PREÇO MÍNIMO DISTRIBUIÇÃO','PREÇO MÁXIMO DISTRIBUIÇÃO','COEF DE VARIAÇÃO DISTRIBUIÇÃO']:
    data_pre[atributo] = pd.to_numeric(data_pre[atributo], errors = 'coerce')
```

```python
data_pre.info()
```

```python
#Criando mascara para identificar dados nulos:
mask = data_pre['PREÇO MÉDIO DISTRIBUIÇÃO'].isnull()
```

```python
data_pre[mask]
```

```python
data[mask]
```

```python
data_pre.dropna(inplace= True)
```

```python
data_pre.info()
```

```python
data_pre.to_csv('./dataset/GasPricesinBrazil_2004-2019_pre_processado.csv', index = False)
```

```python
data_pre
```

```python
data_final = data_pre
```

```python
data_final
```

```python
#Não usaremos os casos de 2019, pois o ano nao esta completo:
data_filtrado = data_final.query('ANO!=2019')
data_filtrado
```

```python
#QUANTOS REGISTROS DE PRODUTOS TEM EM CADA REGIAO?
grupos = data_filtrado.groupby('PRODUTO')
```

```python
grupos
```

```python
grupos['REGIÃO'].value_counts().to_frame()
```

```python
#como os preços da gasolina variaram em SAO PAULO, em 2018?
postos_sp = data_filtrado.query('ESTADO == "SAO PAULO"')
```

```python
postos_sp
```

```python
postos_sp2018 = postos_sp.query('ANO == 2018')
postos_sp2018
```

```python
gasol_sp2018 = postos_sp2018.query('PRODUTO == "GASOLINA COMUM"')
gasol_sp2018
```

```python
gasol_sp2018['PREÇO MÉDIO REVENDA'].describe().to_frame()
```

```python
print(f'A variação no preço Médio de gasolina de São Paulo de 2018, foi de {gasol_sp2018['PREÇO MÉDIO REVENDA'].std():.2f} Reais')
```

```python
#COMO OS PREÇOS DA GASOLINA E ETANOL VARIARAM EM 2018, EM SAO PAULO?
postos_sp2018
```

```python
gasol_e_etanol = postos_sp2018.query('PRODUTO == ["GASOLINA COMUM","ETANOL HIDRATADO"]')
gasol_e_etanol
```

```python
gasol_e_etanol['PREÇO MÉDIO REVENDA'].describe().to_frame()
```

```python
print(f'Neste caso, contando tanto GASOLINA, quanto ETANOL, tivemos uma variação de {gasol_e_etanol['PREÇO MÉDIO REVENDA'].std():.2f} Reais')
```
