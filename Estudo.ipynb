```python
#FILTRAGEM COM MAIS COMPARAÇÕES:
#Aqui vamos selecionar os registros de postos do Rio de Janeiro, com preços acima de 2 reais.
data['ESTADO'] == 'RIO DE JANEIRO'
```

```python
data['PREÇO MÉDIO REVENDA'] > 2.0
```

```python
#para juntarmos as duas series:
#o & funciona como and.
selecaorj = (data['ESTADO'] == 'RIO DE JANEIRO')&( data['PREÇO MÉDIO REVENDA'] > 2.0).reset_index(drop = True)
selecaorj
#Nos retorna se é do Rio de Janeiro E se é maior que 2 reais:
```

```python
# | representa o OR
# ~ representa o NOT
```

```python
data[selecaorj].reset_index(drop = True)
```

```python
#NESTE CASO, COMO A COLUNA DO PREÇO MEDIO DE REVENDA, CONTEM CARACTERES INVALIDOS, DARÁ ERRO E NÃO SERÁ POSSIVEL POR MEIO DE UMA QUERY.
```

```python
#Esta abordagem que estamos usando, é boa até que se tenha um enorme Dataframe, pois, quanto maior, mais linhas a query vai ter que percorrer e será mais lento
#Para otimizar essa busca, faremos da seguinte forma:
```

```python
#Primeiro, iremos reduzir o dataframe, somente ao que queremos analisar
selecao1 = data['ESTADO'] == 'RIO DE JANEIRO'
postos_rj = data[selecao1]
postos_rj
```

```python
#feito isso, reduzimos novamente, à segunda questão que sera analisada
selecao_2 = postos_rj['PREÇO MÉDIO REVENDA'] > 2
selecao_2
```

```python
#apos serem feitas as reduções, podemos adicionar a um novo dataframe, quais as observações solicitadas, sempre diminuindo o tamanho de linhas que o programa deverá percorrer
postos_rj_preco_maior_2 = postos_rj[selecao_2]
postos_rj_preco_maior_2
```
